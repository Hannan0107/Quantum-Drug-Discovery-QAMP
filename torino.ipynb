```python

import numpy as np
import pandas as pd
import requests
import io
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

# ML & Metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from rdkit import Chem
from rdkit.Chem import AllChem

# Quantum
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2 as RuntimeSampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

warnings.filterwarnings('ignore')
print("--- HERO REPLICATION RUN (Restoring Bond Logic) ---")


# 1. THE "HERO" KERNEL

EN_MAP = {'H': 2.20, 'Li': 0.98, 'Be': 1.57, 'B': 2.04, 'C': 2.55, 'N': 3.04, 'O': 3.44, 'F': 3.98, 'S': 2.58, 'Cl': 3.16}

def qmse_hero_normal(smiles):
    angle_scale = 0.1
    n_qubits = 12
    mol = Chem.MolFromSmiles(smiles)
    if not mol: return QuantumCircuit(n_qubits)

    atoms = [a for a in mol.GetAtoms()]
    qc = QuantumCircuit(n_qubits)

    # 1. NORMAL MAPPING (Preserves Geometry)
    n_exact = 10
    backbone = atoms[:n_exact]
    tail = atoms[n_exact:]
    mol_to_qubit = {atom.GetIdx(): i for i, atom in enumerate(backbone)}

    # 2. ENCODING ATOMS
    for atom in backbone:
        en = EN_MAP.get(atom.GetSymbol(), 1.0)
        qc.ry((en**2) * angle_scale, mol_to_qubit[atom.GetIdx()])

    # 3. COMPRESSING TAIL
    if tail:
        tail_A = tail[0::2]
        tail_B = tail[1::2]
        if tail_A: qc.ry(sum(EN_MAP.get(a.GetSymbol(), 1.0) for a in tail_A) * (angle_scale/2), 10)
        if tail_B: qc.ry(sum(EN_MAP.get(a.GetSymbol(), 1.0) for a in tail_B) * (angle_scale/2), 11)

    # 4. ENCODING BONDS
    for bond in mol.GetBonds():
        idx1, idx2 = bond.GetBeginAtomIdx(), bond.GetEndAtomIdx()
        q1, q2 = mol_to_qubit.get(idx1), mol_to_qubit.get(idx2)

        # Use Bond Type (1.0 for Single, 2.0 for Double, 1.5 for Aromatic)
        b_type = bond.GetBondTypeAsDouble()

        if q1 is not None and q2 is not None:
            qc.rxx(b_type * 0.5, q1, q2)
        elif (q1 is not None):
            qc.rxx(0.5, q1, 10); qc.rxx(0.5, q1, 11)
        elif (q2 is not None):
            qc.rxx(0.5, q2, 10); qc.rxx(0.5, q2, 11)

    qc.measure_all()
    return qc

def extract_features(pub_result, n_qubits=12):
    c = pub_result.data.meas.get_counts()
    p = np.zeros(2**n_qubits)
    tot = sum(c.values())
    for k, v in c.items(): p[int(k, 2)] = v/tot
    return p

def print_metrics_table(name, y_true, y_pred, y_prob):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    auc_val = roc_curve(y_true, y_prob)[1][1] if len(np.unique(y_true)) > 1 else 0.5
    print(f"    {name:10} | Acc: {acc:.1%} | Prec: {prec:.2f} | Rec: {rec:.2f} | F1: {f1:.2f} | AUC: {auc_val:.2f}")

def is_valid_mol(s):
    try: return Chem.MolFromSmiles(s) is not None
    except: return False


# 2. SETUP & DATASETS

datasets = {}

# Load BBBP
try:
    print("\n1. Loading BBBP...")
    df = pd.read_csv("https://deepchemdata.s3-us-west-1.amazonaws.com/datasets/BBBP.csv")
    df['target'] = df['p_np']
    df['SMILES'] = df['smiles']
    datasets['BBBP'] = df
except: print("   BBBP Load Failed")

# Load B3DB
try:
    print("2. Loading B3DB...")
    s = requests.get("https://raw.githubusercontent.com/theochem/B3DB/main/B3DB/B3DB_classification.tsv").content
    df = pd.read_csv(io.StringIO(s.decode('utf-8')), sep='\t')
    df['target'] = df['BBB+/BBB-'].map({'BBB+': 1, 'BBB-': 0})
    datasets['B3DB'] = df
except: print("    B3DB Load Failed")

# Auth
token = "hBz11eVG8IHUHKaQ559UldN2YA1azgN_DOc7CzfwK1QF"
instance = "crn:v1:bluemix:public:quantum-computing:us-east:a/9056bc7c796c4a32927d5273205648dc:80af8a07-ac2e-4fa6-8287-56cf407151ee::"

try:
    QiskitRuntimeService.save_account(
        token=token,
        channel="ibm_cloud",
        instance=instance,
        name="Soko",
        overwrite=True
    )
except:
    pass # Account might already be saved
service = QiskitRuntimeService(name="Soko")
backend = service.backend("ibm_torino")
print(f"✅ Target Backend: {backend.name}")

results_store = {}


# 3. EXECUTION LOOP

for name, df in datasets.items():
    print(f"\n" + "="*60)
    print(f" PROCESSING: {name} (Method: Normal + Bond Types)")
    print("="*60)

    # A. PREPROCESS (Exact 250 Total -> 200 Train / 50 Test)
    df = df.dropna(subset=['target', 'SMILES'])
    df = df[df['SMILES'].apply(is_valid_mol)]
    df = df[df['SMILES'].str.len() < 70]

    df_pos = df[df['target'] == 1]
    df_neg = df[df['target'] == 0]

    n_sub = 125 # 125 + 125 = 250 Total

    if len(df_pos) < n_sub or len(df_neg) < n_sub:
        print(f"Not enough data for {name}. Skipping.")
        continue

    df_final = pd.concat([df_pos.sample(n_sub, random_state=42), df_neg.sample(n_sub, random_state=42)]).sample(frac=1, random_state=42)
    y = df_final['target'].values
    smiles_list = df_final['SMILES'].values
    print(f"   Data Ready: {len(df_final)} molecules (Total 250).")

    # B. GENERATE CIRCUITS (Using Hero Function)
    print("   1. Generating Circuits (with Bond Logic)...")
    circuits = [qmse_hero_normal(s) for s in smiles_list]

    # C. HARDWARE EXECUTION
    print("   2. Submitting to Hardware (1024 Shots)...")
    X_q_hw = None
    try:
        pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
        isa_circuits = pm.run(circuits)

        sampler_hw = RuntimeSampler(backend)
        sampler_hw.options.default_shots = 1024 # Restored to 1024

        # Using PUBs for batching
        pubs = [(c,) for c in isa_circuits]

        job = sampler_hw.run(pubs)
        print(f"      Job ID: {job.job_id()} (Waiting...)")
        result_hw = job.result()

        X_q_hw = []
        for pub in result_hw: X_q_hw.append(extract_features(pub))
        X_q_hw = np.array(X_q_hw)
        print("  Hardware Data Received.")
    except Exception as e:
        print(f" Hardware Failed: {e}")

    # D. SCORING (200 Train / 50 Test)
    if X_q_hw is not None:
        # Use Classical Fingerprints for Baseline comparison
        X_c = np.array([np.array(AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(s), 2, nBits=256)) for s in smiles_list])

        # 0.2 of 250 = 50 Test Samples
        sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
        train_idx, test_idx = next(sss.split(X_c, y))

        print(f"\n  Split Info: Train Size = {len(train_idx)}, Test Size = {len(test_idx)}")
        print("\n   PERFORMANCE METRICS:")

        # Classical
        rf = RandomForestClassifier(n_estimators=100); rf.fit(X_c[train_idx], y[train_idx])
        res_c = (y[test_idx], rf.predict(X_c[test_idx]), rf.predict_proba(X_c[test_idx])[:, 1])
        print_metrics_table("Classical", *res_c)

        # Hardware
        sc_h = StandardScaler()
        svc_h = SVC(kernel='rbf', C=10, probability=True)
        svc_h.fit(sc_h.fit_transform(X_q_hw[train_idx]), y[train_idx])
        res_h = (y[test_idx], svc_h.predict(sc_h.transform(X_q_hw[test_idx])), svc_h.predict_proba(sc_h.transform(X_q_hw[test_idx]))[:, 1])
        print_metrics_table("Hardware", *res_h)

        results_store[name] = {'Classical': res_c, 'Hardware': res_h}


# 4. PLOTTING

print("\n" + "="*60)
print("GENERATING PLOTS")
print("="*60)

for name, res_dict in results_store.items():
    plt.figure(figsize=(14, 6))

    # ROC
    plt.subplot(1, 2, 1)
    colors = {'Classical': 'black', 'Hardware': 'red'}
    for model, (yt, yp, yprob) in res_dict.items():
        fpr, tpr, _ = roc_curve(yt, yprob)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{model} (AUC={roc_auc:.2f})", color=colors.get(model, 'green'), linewidth=2)
    plt.plot([0,1],[0,1], 'k:')
    plt.title(f"{name}: ROC Curve")
    plt.legend()
    plt.grid(True, alpha=0.3)

    # CM
    plt.subplot(1, 2, 2)
    cm = confusion_matrix(res_dict['Hardware'][0], res_dict['Hardware'][1])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', cbar=False,
                xticklabels=['Neg', 'Pos'], yticklabels=['Neg', 'Pos'])
    plt.title(f"{name}: Confusion Matrix (Hardware)")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")

    plt.tight_layout()
    plt.show()

print("\n EXPERIMENT COMPLETE.")
```
